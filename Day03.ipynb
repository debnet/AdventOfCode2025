```python
with open("Day03.txt") as file:
    banks = file.read().splitlines()
```

```python
def get_max_joltage(bank, count = 12):
    result, lower, upper = "", 0, len(bank) - count + 1
    for _ in range(count):
        chunk = bank[lower:upper]
        best = max(chunk)
        result += best
        lower += chunk.index(best) + 1
        upper += 1
    return int(result)
```

```python
sum(get_max_joltage(bank, count=2) for bank in banks)
```

```python
sum(get_max_joltage(bank, count=12) for bank in banks)
```
