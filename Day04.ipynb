```python
with open("Day04.txt") as file:
    data = file.read()

grid = set()
for x, line in enumerate(data.splitlines()):
    for y, char in enumerate(line):
        if char == "@":
            grid.add((x, y))
width, height = x + 1, y + 1
```

```python
def neighbors(x: int, y: int):
    result = set()
    for j in (-1, 0, +1):
        for i in (-1, 0, +1):
            if (i, j) == (0, 0):
                continue
            p = (x + i, y + j)
            if p in grid:
                result.add(p)
    return result
```

```python
prev = set((x, y) for x, y in grid if len(neighbors(x, y)) < 4)
len(prev)
```

```python
count = len(prev)
grid -= prev
while prev:
    prev = set((x, y) for x, y in grid if len(neighbors(x, y)) < 4)
    count += len(prev)
    grid -= prev
count
```
