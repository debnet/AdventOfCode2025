```python
with open("Day12.txt") as file:
    data = file.read()
```

```python
import re

regex_boards = re.compile(r"(\d+)x(\d+): (\d+) (\d+) (\d+) (\d+) (\d+) (\d+)")
boards = [tuple(map(int, board)) for board in regex_boards.findall(data)]
sum(w // 3 * h // 3 >= sum(n) for w, h, *n in boards)
```
