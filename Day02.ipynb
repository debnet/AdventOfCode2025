```python
def irange(r: str):
    a, b = map(int, r.split("-"))
    return range(a, b + 1)

with open("Day02.txt") as file:
    ranges = [irange(r) for r in file.read().split(",")]
```

```python
t = []
for r in ranges:
    for n in r:
        s = str(n)
        l = len(s) // 2
        if s[:l] == s[l:]:
            t.append(n)
sum(t)
```

```python
from itertools import batched

t = []
for r in ranges:
    for n in r:
        s = str(n)
        for i in range(1, (len(s) // 2) + 1):
            if len(set(batched(s, i))) == 1:
                t.append(n)
                break
sum(t)
```
