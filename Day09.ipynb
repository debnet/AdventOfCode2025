```python
with open("Day09.txt") as file:
    points = [tuple(map(int, p.split(","))) for p in file.read().split()]
```

```python
from itertools import combinations

def area(a, b):
    (x1, y1), (x2, y2) = a, b
    return (abs(x1 - x2) + 1) * (abs(y1 - y2) + 1)

max(area(a, b) for a, b in combinations(points, 2))
```

```python
from shapely import Polygon
from shapely.geometry import box

def zone(a, b):
    (x1, y1), (x2, y2) = a, b
    (xmin, xmax), (ymin, ymax) = sorted((x1, x2)), sorted((y1, y2))
    return box(xmin, ymin, xmax, ymax)

tiles = Polygon(points)
max(area(a, b) for a, b in combinations(points, 2) if zone(a, b).within(tiles))
```
