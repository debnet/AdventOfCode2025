```python
with open("Day05.txt") as file:
    ranges, ingredients = file.read().split("\n\n")

ranges, ingredients = (
    sorted(tuple(map(int, r.split("-"))) for r in ranges.splitlines()),
    sorted(map(int, ingredients.splitlines()))
)
```

```python
sum(any(a <= i <= b for a, b in ranges) for i in ingredients)
```

```python
count, pa, pb = 0, 0, 0
for a, b in ranges:
    a = max(a, pb)
    b = max(a, b)
    count += b - a + (a > pb)
    pa, pb = a, b
count
```
