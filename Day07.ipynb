```python
with open("Day07.txt") as file:
    grid = {(x, y): c for y, r in enumerate(file.read().split()) for x, c in enumerate(r)}
start = next((x, y) for (x, y), c in grid.items() if c == "S")
```

```python
from collections import defaultdict, deque

todo, seen, logs, splits = deque([start]), {start}, defaultdict(int, {start: 1}), 0
while todo:
    x, y = todo.popleft()
    nxt = (x, y + 1)
    if nxt in seen or nxt not in grid:
        continue
    seen.add(nxt)
    if grid[nxt] == "^":
        s1, s2 = (x - 1, y + 1), (x + 1, y + 1)
        todo.extend((s1, s2))
        logs[s1] += logs[x, y]
        logs[s2] += logs[x, y]
        splits += 1
    else:
        todo.append(nxt)
        logs[nxt] += logs[x, y]
```
