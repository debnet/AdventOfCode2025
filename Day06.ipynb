```python
from math import prod

with open("Day06.txt") as file:
    data = file.read().splitlines()

operators = list(map({"+": sum, "*": prod}.get, data[-1].split()))
```

```python
columns = [line.split() for line in data[:-1]]
columns = [[int(line[index]) for line in columns] for index in range(len(operators))]
sum(operator(column) for column, operator in zip(columns, operators))
```

```python
chars = [list(line) for line in data[:-1]]
columns = list("".join(column).strip() for column in zip(*chars))
columns = [[int(number) for number in column.split(",")] for column in ",".join(columns).split(",,")]
sum(operator(column) for column, operator in zip(columns, operators))
```
