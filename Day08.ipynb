```python
with open("Day08.txt") as file:
    data = file.read()

boxes = [tuple(map(int, box.split(","))) for box in data.split()]
```

```python
from itertools import combinations

def get_distance(a, b):
    (x1, y1, z1), (x2, y2, z2) = a, b
    return (x1 - x2) ** 2 + (y1 - y2) ** 2 + (z1 - z2) ** 2

distances = sorted((get_distance(a, b), a, b) for a, b in combinations(boxes, 2))
```

```python
circuits, limit = [], 1000
for d, a, b in distances:
    circuit = {a, b}
    for c in circuits[:]:
        if {a, b} & c:
            circuit |= c
            circuits.remove(c)
    circuits.append(circuit)
    limit -= 1
    if not limit:
        best = circuits[:]
    if len(circuits) == 1 and len(circuits[0]) == len(boxes):
        break
```

```python
from math import prod

prod(sorted((len(c) for c in best), reverse=True)[:3])
```
