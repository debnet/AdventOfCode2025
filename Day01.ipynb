```python
with open("Day01.txt") as file:
    lines = file.read().split()
directions =  [(-1, +1)[l[0] == "R"] for l in lines]
clicks = [int(l[1:]) for l in lines]
```

```python
position, previous = 50, +1
score_p1, score_p2 = 0, 0
for direction, click in zip(directions, clicks):
    if previous != direction:
        position = 100 - position
        previous = direction
    position = (position % 100) + click
    score_p1 += (position % 100 == 0)
    score_p2 += position // 100
```
