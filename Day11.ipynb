```python
with open("Day11.txt") as file:
    graph = {k.removesuffix(":"): set(v) for k, *v in map(str.split, file.readlines())}
```

```python
from functools import cache

@cache
def count(src, dst, **cfg):
    if src in cfg:
        cfg[src] = True
    elif src == dst:
        return all(cfg.values())
    return sum(count(nxt, dst, **cfg) for nxt in graph[src])
```

```python
count("you", "out")
```
